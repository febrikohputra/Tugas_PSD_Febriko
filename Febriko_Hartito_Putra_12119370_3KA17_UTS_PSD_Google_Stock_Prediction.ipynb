#Nomor 3
Memilih dan mencetak hasil prediksi pada tanggal 29 September 2021

```python
df = df.head(len(df)-1)
df
```

Menyiapkan data untuk pelatihan. Membuat ulang tabel data dengan mendapatkan semua data kecuali untuk baris terakhir untuk menguji model tersebut dan menyimpan data baru dengan baris terakhir hilang kembali ke 'df'.

```python
#Create the lists / X and y data set
days = list()
adj_close_prices = list()
```

Membuat variabel yang akan digunakan sebagai kumpulan data independen dan dependen dengan mengaturnya sama dengan daftar kosong.

```python
df_days = df.loc[:, 'Date']
df_adj_close = df.loc[:, 'Adj Close']
```

Mendapatkan semua baris dari kolom tanggal simpan ke dalam variabel yang disebut 'df_days' dan mendapatkan semua baris dari kolom Harga Tutup Adj dan menyimpan data ke dalam variabel.

```python
#Create the independent data set
for day in df_days:
   days.append( [int(day.split('-')[2])] )
#Create the dependent data set
for adj_close_price in df_adj_close:
   adj_close_prices.append( float(adj_close_price) )
```

Untuk mengumpulkan data independen, yang diinginkan hari dari tanggal tersebut, jadi menggunakan fungsi split untuk mendapatkan hari dan mentransmisikannya ke bilangan bulat sambil menambahkan data ke daftar hari.

```python
print(days)
```

Melihat hari apa saja yang ada pada dataset

```python
#Create and train an SVR model using a linear kernel
lin_svr = SVR(kernel='linear', C=1000.0)
lin_svr.fit(days,adj_close_prices)#Create and train an SVR model using a polynomial kernel
poly_svr = SVR(kernel='poly', C=1000.0, degree=2)
poly_svr.fit(days, adj_close_prices)#Create and train an SVR model using a RBF kernel
rbf_svr = SVR(kernel='rbf', C=2500.0, gamma=0.20)
rbf_svr.fit(days, adj_close_prices)
```

Membuat dan melatih 3 model Support Vector Regression (SVR) yang berbeda dengan tiga kernel berbeda untuk melihat mana yang berkinerja terbaik.

```python
#Plot the models on a graph to see which has the best fit
plt.figure(figsize=(16,8))
plt.scatter(days, adj_close_prices, color = 'black', label='Original Data')
plt.plot(days, rbf_svr.predict(days), color = 'green', label='RBF Model')
plt.plot(days, poly_svr.predict(days), color = 'orange', label='Polynomial Model')
plt.plot(days, lin_svr.predict(days), color = 'purple', label='Linear Model')
plt.xlabel('Days')
plt.ylabel('Adj Close Price')
plt.title('Support Vector Regression')
plt.legend()
plt.show()
```

```python
NOMOR 4
Model yang paling baik untuk menentukan prediksi pada tanggal 29 September 2021 adalah RBF. Karena, RBF merupakan model Support Vector Regression yang menggunakan kernel yang disebut fungsi basis radial.
```

```python
day = [[30]]
print('The RBF SVR predicted:', rbf_svr.predict(day))
print('The Linear SVR predicted:', lin_svr.predict(day))
print('The Polynomial SVR predicted:', poly_svr.predict(day))
```
